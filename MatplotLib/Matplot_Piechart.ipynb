```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("F:/python_workspace/CodeBasics/MatplotLib/data/company_sales_data.csv")
```

```python
df.head()
```

```python
column_list = df.columns
column_list = column_list.delete(0)
column_list = column_list.delete(7)
column_list = column_list.delete(6)
```

```python
avgPro_sale_list = []
avgPro_sale_list = map(lambda x: int(df[x].describe().mean()) ,column_list)
avgPro_sale_list = list(avgPro_sale_list)
avgPro_sale_list
```

```python
plt.axis('equal')
plt.pie(list(avgPro_sale_list),labels = column_list, radius=1.5, autopct='%0.2f%%',)
plt.show()
```

```python
plt.axis('equal')
plt.pie(list(avgPro_sale_list),labels = column_list, radius=1.5, autopct='%0.1f%%',explode=[0,0.2,0,0,0.2,0],startangle=45)
plt.show()
```

# Save Figure In File

```python
plt.axis('equal')
plt.pie(list(avgPro_sale_list),labels = column_list, radius=1.5, autopct='%0.1f%%',explode=[0,0.2,0,0,0.2,0],startangle=45)
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=2,transparent=True)
```
