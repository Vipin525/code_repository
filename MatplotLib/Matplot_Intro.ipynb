```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df =pd.read_csv("F:/python_workspace/CodeBasics/MatplotLib/data/company_sales_data.csv")
```

```python
df.head()
```

```python
#Introduction of matplotlib
plt.xlabel("Month")
plt.ylabel("Facecream Sales Counts")
plt.title("Company Sales Report")
plt.plot(df['month_number'],df['facecream'],color='Red')
```

```python
# Format String in 'plot' function
# Matplotlib api for plot -- https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
#what is 'g+--' -- it means 'g'=Green color, '+'= Marker and '--' = connecting component it simmilar to 
#plt.plot(x,y,color='green',marker='+',linestyle='--')
plt.xlabel('Month')
plt.ylabel('Facewash')
plt.title('Company Sales Chart')
plt.plot(df['month_number'],df['facewash'],'g+--')
```

```python
plt.plot(df['month_number'],df['facewash'],'g+',markersize=10)
```

```python
plt.plot(df['month_number'],df['facewash'],'g+--',alpha=0.2)
```

```python
#Introduction of Axes labels, legend and grid
plt.plot(df['month_number'],df['facecream'], label='Facecream')
plt.plot(df['month_number'],df['facewash'], label='Facewash')
plt.plot(df['month_number'],df['toothpaste'],label='Toothpaste')
plt.plot(df['month_number'],df['bathingsoap'],label='Bath Soap')
plt.plot(df['month_number'],df['shampoo'],label='Shampoo')
#plt.legend(loc='lower left')
plt.legend(loc='best', shadow=True, fontsize='x-small') # 'Best' will find the suitable position for legend on graph
plt.grid()
```
