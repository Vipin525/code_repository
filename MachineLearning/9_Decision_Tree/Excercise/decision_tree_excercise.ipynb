```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
feature_input = df.drop(['Pclass','PassengerId','Name','Survived','SibSp','Ticket','Cabin','Embarked'],axis= 'columns')
target_input = df['Survived']
```

```python
feature_input.head()
```

```python
feature_input.Age = feature_input.Age.fillna(feature_input.Age.mean())
```

```python
feature_input
```

```python
le = LabelEncoder()
```

```python
feature_input['Sex_le'] = le.fit_transform(feature_input['Sex'])
```

```python
feature_input = feature_input.drop(['Sex'],axis='columns')
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_input,target_input,test_size=0.1)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
